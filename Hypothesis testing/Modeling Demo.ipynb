```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.polynomial import polynomial
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error    
```

# Modeling Demo

## First steps

```python
x = np.linspace(-5, 5, 1000)
y = 2 * x + 3
y_noise = np.random.normal(0, 1.8, 1000)
y += y_noise
```

```python
plt.scatter(x, y, alpha=0.5, s=1)
```

```python
a = np.random.choice(np.linspace(-3, 3, 1000))
b = np.random.choice(np.linspace(-3, 3, 1000))
```

```python
std = np.random.choice(np.linspace(0.1, 2, 20))
```

```python
x = np.linspace(-5, 5, 1000)
y = a * x + b
y_noise = np.random.normal(0, std, 1000)
y += y_noise
```

```python
plt.scatter(x, y, alpha=0.5, s=1)
```

```python
# Assumptions:
#1. y = f(x)
#2. y = ax + b; a and b are unknown _parameters_ \in \matthbb{R}
#3. noise we are not interested in, but it is there and it is normally distributed with mean 0 and some unknown standard deviation 
```

```python
model = lambda x: 5 * x + 2
```

```python
model(x)
```

```python
def model(data):
    return 5 * data + 2
```

```python
predictions = model(x)
```

```python
plt.scatter(x, y, alpha=0.5, s=1)
plt.plot(x, predictions, color='red', label = "prediction")
plt.legend()
plt.xlabel("x")
plt.ylabel("y") 
plt.show()
```

```python
def model(data):
    return -2 * data + 1

predictions = model(x)

plt.scatter(x, y, alpha=0.5, s=1)
plt.plot(x, predictions, color='red', label = "prediction")
plt.legend()
plt.xlabel("x")
plt.ylabel("y") 
plt.show()
```

```python
def model(data):
    return 0.5 * data + 1

predictions = model(x)

plt.scatter(x, y, alpha=0.5, s=1)
plt.plot(x, predictions, color='red', label = "prediction")
plt.legend()
plt.xlabel("x")
plt.ylabel("y") 
plt.show()
```

```python
def model(data, a = 0, b = 0):
    return a * data + b

predictions = model(x, -1, 2)

plt.scatter(x, y, alpha=0.5, s=1)
plt.plot(x, predictions, color='red', label = "prediction")
plt.legend()
plt.xlabel("x")
plt.ylabel("y") 
plt.show()
```

```python
np.abs(predictions - y).mean()
```

```python
def mae(y, prediction):
    return np.abs(prediction - y).mean()
```

```python
predictions = model(x, -1, 2)
mae(y, predictions)
```

```python
polynomial.polyfit(x, y, deg=1)
```

```python
a_model, b_model = polynomial.polyfit(x, y, deg=1)
```
