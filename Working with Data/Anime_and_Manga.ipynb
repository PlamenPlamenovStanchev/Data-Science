```python
merged = anime.merge(manga, on="title", how="inner").drop_duplicates("title")
```

```python
plt.scatter(merged.score_x, merged.score_y)
plt.xlabel('Anime Score')
plt.ylabel('Manga Score')
plt.title('Anime vs Manga Scores')  
plt.show()
```

```python
plt.scatter(merged.score_x, merged.score_y, s=1)
plt.xlabel('Anime Score')
plt.ylabel('Manga Score')
plt.title('Anime vs Manga Scores')  
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.gca().set_aspect('equal')
plt.show()
```
