### Problem 4. Missing values
Try to distinguish between missing values and data that simply does not exist. How will you treat them? Execute your plans that you should already have from the previous problem. How about anomalous values (anomalies / outliers / novelties)?

```python
# categorical columns
categorical_cols = ['type', 'status', 'rating', 'source', 'sfw', 'demographics']
for col in categorical_cols:
    if col in anime.columns:
        anime[col] = anime[col].fillna('Unknown')
    if col in manga.columns:
        manga[col] = manga[col].fillna('Unknown')

# text columns
text_cols = ['synopsis', 'genres', 'themes', 'studios', 'producers', 'licensors', 'authors', 'serializations']
for col in text_cols:
    if col in anime.columns:
        anime[col] = anime[col].fillna('[]')
    if col in manga.columns:
        manga[col] = manga[col].fillna('[]')

# numeric columns such as score. We will not use fillna yet because missing score could be meaningful
print("Missing values in anime treated. Remaining nulls:\n", anime.isnull().sum()[anime.isnull().sum() > 0])
print("\nMissing values in manga treated. Remaining nulls:\n", manga.isnull().sum()[manga.isnull().sum() > 0])
```

### Problem 7. Lists, part 1: descriptions
Many of the columns are not tidy at all. They contain an entire list within a cell. What can you do about them?

Start by providing descriptive statistics about such columns. How long are these lists? How many of them only have one element; e.g., for genres - how useful would it be to only consider one genre per anime / manga? How useful are manga demongraphics and how often do they appear in sets of many? Try exploring the data and maybe applying some domain knowledge. Don't forget to analyze empty lists.

```python
import ast

def parse_list(x):
    if pd.isna(x):
        return []
    try:
        parsed = ast.literal_eval(x)
        if isinstance(parsed, list):
            return parsed
        return []
    except (ValueError, SyntaxError):
        return []

anime['genres_list'] = anime['genres'].apply(parse_list)
anime['genres_count'] = anime['genres_list'].apply(len)

manga['genres_list'] = manga['genres'].apply(parse_list)
manga['genres_count'] = manga['genres_list'].apply(len)

print("Number of genres per anime distribution:")
print(anime['genres_count'].value_counts().sort_index())

print("\nNumber of genres per manga distribution:")
print(manga['genres_count'].value_counts().sort_index())

anime['genres_count'].hist(bins=range(12), align='left', rwidth=0.8, alpha=0.7, label='Anime')
manga['genres_count'].hist(bins=range(12), align='left', rwidth=0.8, alpha=0.7, label='Manga')
plt.title('Number of Genres')
plt.xlabel('Number of Genres')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Problem 11. Support / popularity
Let's revisit the questions we asked earlier (and left). Find some highly rated works, studios, authors, sources, or categories.

Now revisit your results and ask whether the ranking is still convincing once you take support into account. A title with a sky-high score and very few scorers is not quite the same creature as one with a slightly lower score and a massive audience.

You may define "support" in a way you find reasonable, but explain it.

```python
# Let's calculate a Bayesian average to take support (members) into account
# C is the mean of members across the dataset
# m is the mean score across the dataset
C_anime = anime['members'].mean()
m_anime = anime['score'].mean()

def bayesian_rating(v, R, C, m):
    if pd.isna(R) or pd.isna(v):
        return np.nan
    return (v / (v + C)) * R + (C / (v + C)) * m

anime['bayesian_score'] = anime.apply(lambda row: bayesian_rating(row['members'], row['score'], C_anime, m_anime), axis=1)

print("Top 10 True Best Anime (Bayesian Score):")
display(anime.sort_values('bayesian_score', ascending=False)[['title', 'score', 'bayesian_score', 'members']].head(10))

# The same logic for studios based on total true score of their exploded views could be applied
```

### Problem 12. Anime versus manga

Compare the two datasets. What do they have in common? A very obvious place to start is comparing the names as sets.

Which columns seem similar? Which are medium-specific? Which features can reasonably be compared, and which ones should not be?

Can you create a unified dataset of works? Why / why not? How? What will it include and why?

```python
print("Anime columns:", set(anime.columns))
print("Manga columns:", set(manga.columns))

common_cols = list(set(anime.columns).intersection(set(manga.columns)))
print("\nCommon columns:", common_cols)

# We can find matches based on title
common_titles = set(anime['title']).intersection(set(manga['title']))
print(f"\nNumber of shared titles: {len(common_titles)}")

# We can create a unified dataset
anime['medium'] = 'anime'
manga['medium'] = 'manga'

unified_works = pd.concat([anime[common_cols + ['medium']], manga[common_cols + ['medium']]], ignore_index=True)
print(f"\nUnified dataset shape: {unified_works.shape}")
display(unified_works.head())
```

### * Problem 13. Text processing
Try to normalize names, tags, etc. Think about whitespace, word forms, etc. You'll find yourself dealing with both English and Japanese, so be careful about the encoding (the columns should be Unicode-encoded) and special symbols / numbers / etc.

Perform any text processing that you might see fit.

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text)
    # Remove MAL writeup credits
    text = re.sub(r'\[Written by MAL Rewrite\]', '', text)
    # Remove extra whitespaces and line breaks
    text = re.sub(r'\s+', ' ', text).strip()
    return text

anime['synopsis_clean'] = anime['synopsis'].apply(clean_text)
manga['synopsis_clean'] = manga['synopsis'].apply(clean_text)

print("Original Synopsis excerpt:")
print(anime.loc[0, 'synopsis'][:200])

print("\nCleaned Synopsis excerpt:")
print(anime.loc[0, 'synopsis_clean'][:200])
```
